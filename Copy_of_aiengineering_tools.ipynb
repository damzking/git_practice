```python
# -----------------------------
# 1) Install dependencies (run once in a notebook)
# -----------------------------
!pip install -q sentence-transformers rank-bm25 scikit-learn numpy


# -----------------------------
# 2) Imports
# -----------------------------
from rank_bm25 import BM25Okapi                       # BM25 keyword-based search
from sentence_transformers import SentenceTransformer, util  # embeddings + cosine similarity
import numpy as np                                   # numerical operations


# -----------------------------
# 3) Documents (Your Knowledge Base)
# -----------------------------
docs = [
    "Enable two-factor authentication (2FA) in your account settings to add an extra security step.",
    "HbA1c measures long-term glucose; talk to your physician about tests for glycated hemoglobin.",
    "Our PTO policy covers paid time off for vacations and sick leave.",
    "How to fix engine misfires caused by bad spark plugs.",
    "Kubernetes Ingress configuration for path-based routing.",
    "Configure MFA with authenticator apps.",
    "Doctor appointment scheduling policy."
]

# Example queries we want to search for
queries = [
    "How do I set up 2FA?",
    "What does HbA1c mean?",
    "sick leave policy"
]


# -----------------------------
# 4) BM25 Setup (Keyword Search)
# -----------------------------
tokenized_corpus = [d.lower().split() for d in docs]
# Convert each document into lowercase tokens (words)
# Example: "Hello World" -> ["hello", "world"]

bm25 = BM25Okapi(tokenized_corpus)
# Build BM25 search index using the tokenized documents


# -----------------------------
# 5) Sentence Transformer Setup (Semantic Search)
# -----------------------------
model = SentenceTransformer("all-MiniLM-L6-v2")
# Load a pretrained embedding model
# all-MiniLM-L6-v2 is fast and general-purpose

doc_emb = model.encode(
    docs,
    convert_to_tensor=True,          # output as tensor (needed for util.cos_sim)
    normalize_embeddings=True        # normalize embeddings for stable cosine similarity
)
# Create embeddings for all documents once
# This is like "indexing" the docs for semantic search


# -----------------------------
# 6) Search Function (BM25 + Semantic + Hybrid)
# -----------------------------
def show_results(query, k=3, alpha=0.5):
    """
    query: the user search string
    k: number of top results to show
    alpha: weight for BM25 vs semantic
           alpha=0.5 means equal weight
           alpha=0.7 means BM25 matters more
           alpha=0.3 means semantic matters more
    """

    print(f"\nQUERY: {query}\n" + "=" * 60)
    # Print query and separator line


    # -----------------------------
    # A) BM25 Retrieval
    # -----------------------------
    bm25_scores = bm25.get_scores(query.lower().split())
    # Get BM25 relevance score for every document
    # Output is an array of scores with length = number of docs

    top_bm25 = np.argsort(-bm25_scores)[:k]
    # argsort gives indexes that would sort the array
    # -bm25_scores means sort from highest to lowest
    # [:k] means keep top k results

    print("BM25 top-k:")
    # Print label for BM25 results

    for i in top_bm25:
        # Loop through the top BM25 document indexes
        print(f"  [{i}] ({bm25_scores[i]:.3f}) {docs[i]}")
        # Print doc index, BM25 score, and doc text


    # -----------------------------
    # B) Semantic Retrieval (Embeddings)
    # -----------------------------
    q_emb = model.encode(
        query,
        convert_to_tensor=True,         # output tensor
        normalize_embeddings=True       # normalize query embedding too
    )
    # Convert query into an embedding vector

    cos = util.cos_sim(q_emb, doc_emb)[0].cpu().numpy()
    # Compute cosine similarity between query embedding and all doc embeddings
    # util.cos_sim output shape is (1, num_docs)
    # [0] converts it into shape (num_docs,)
    # cpu().numpy() converts tensor -> numpy array

    top_st = np.argsort(-cos)[:k]
    # Sort cosine similarity scores from highest to lowest
    # Get top k document indexes

    print("\nSemantic (SentenceTransformer) top-k:")
    # Print label for semantic results

    for i in top_st:
        # Loop through the semantic top k doc indexes
        print(f"  [{i}] ({cos[i]:.3f}) {docs[i]}")
        # Print doc index, similarity score, and doc text


    # -----------------------------
    # C) Normalize Scores (Needed for Hybrid Fusion)
    # -----------------------------
    bm25_norm = (bm25_scores - bm25_scores.min()) / (
        bm25_scores.max() - bm25_scores.min() + 1e-9
    )
    # Convert BM25 scores into range 0 \u2192 1
    # 1e-9 prevents division by zero if max == min

    st_norm = (cos - cos.min()) / (
        cos.max() - cos.min() + 1e-9
    )
    # Convert semantic cosine scores into range 0 \u2192 1
    # Also uses 1e-9 to avoid division by zero


    # -----------------------------
    # D) Hybrid Fusion
    # -----------------------------
    hybrid = alpha * bm25_norm + (1 - alpha) * st_norm
    # Weighted combination of both systems
    # alpha controls how much BM25 matters

    top_h = np.argsort(-hybrid)[:k]
    # Sort hybrid scores from highest to lowest
    # Get top k indexes

    print("\nHybrid (BM25 + Semantic) top-k:")
    # Print label for hybrid results

    for i in top_h:
        # Loop through top hybrid doc indexes
        print(f"  [{i}] ({hybrid[i]:.3f}) {docs[i]}")
        # Print doc index, hybrid score, and doc text


    # -----------------------------
    # E) Return scores (optional)
    # -----------------------------
    return bm25_scores, cos, hybrid
    # Return raw BM25 scores, semantic scores, and hybrid scores
    # Useful for debugging or plotting later


# -----------------------------
# 7) Run the function on each query
# -----------------------------
for q in queries:
    show_results(q, k=3, alpha=0.5)
    # k=3 shows top 3 results
    # alpha=0.5 means equal weight BM25 and semantic
```
